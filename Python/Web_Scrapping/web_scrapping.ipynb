## Web Scrapping

```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://www.scrapethissite.com/pages/")
```

```python
print(web.status_code)
print(web.content)
```

```python
soup = BeautifulSoup(web.content, "html.parser")
print(soup.prettify())
```

### Tag Get

```python
print(soup.title)
print(soup.title.name)
print("First paragraph tag:",soup.p)
print("First ancher tag:",soup.a)
```

### NavigableString

```python
tag = soup.html
print("First paragraph tag's Sting:",soup.p.string.strip())
```

### BeautifulSoup

```python
print("find first h3: ",soup.find("h3"))
print("find all h3 function: ",soup.find_all("h3"))
```

### comments

```python
com = soup.b
print(com)
```

### prettify

```python
print("Only Body Pretify: \n",soup.body.prettify())
```

### Finding Element By Class

```python
class_data = soup.find_all("div", class_="page")

i = 0
for div in class_data:
    page_title = div.find_all("h3", class_="page-title")
    page_description = div.find_all("p", class_="lead session-desc")
    for title, desc in zip(page_title, page_description):  # zip() :- it can combine any number of iterables: lists, tuples, sets, strings… whatever.
        i=i+1
        print(f"{i}.This is page title:- {title.text.strip()} \n  This page desc:- {desc.text.strip()}")
```

### Extract all the links

```python
for i in soup.find_all("a"):
    print(i.get("href"))
```

### Image Get

```python
image = soup.find_all("img")
for i in image:
    print("Image urls:- ",i.get("src"))
```
